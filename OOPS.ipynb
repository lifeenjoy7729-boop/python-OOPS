```python
Part A: Theory Questions (OOP Concepts)
The following section addresses 25 theoretical questions on Python Object-Oriented Programming concepts with comprehensive explanations.

Q1. What is Object-Oriented Programming (OOP)?

Answer:
Object-Oriented Programming (OOP) is a programming paradigm based on the concept of "objects," which encapsulate data (attributes) and behavior (methods). Unlike procedural programming, where the focus is on functions, OOP models software as a collection of interacting objects. Python is an OOP language that supports all major OOP principles.
Key benefits of OOP include modularity, reusability, scalability, and easier maintenance. Real-world entities (like a Car, BankAccount, or Student) can be directly modeled as classes in OOP.


Q2. What is a class in OOP?

Answer:
A class is a blueprint or template for creating objects. It defines the structure (attributes) and behavior (methods) that its objects will have. In Python, a class is defined using the 'class' keyword.
For example, a class 'Car' may have attributes like brand and model, and methods like start() or stop(). Every individual car created from this class is called an object or instance. Classes help avoid code repetition by centralizing definitions.


Q3. What is an object in OOP?

Answer:
An object is a real-world instance of a class. It is a specific entity created from the class blueprint and possesses its own copy of the attributes defined in the class, along with access to class methods.
For example, if 'Car' is a class, then my_car = Car("Toyota", "Corolla") creates an object named my_car with specific attribute values. Each object is independent; changing one object's attribute does not affect another object of the same class.


Q4. What is the difference between abstraction and encapsulation?

Answer:
Abstraction and Encapsulation are two of the four fundamental OOP pillars, but they serve different purposes:
Abstraction refers to hiding the implementation details and showing only the necessary interface to the user. It focuses on what an object does rather than how it does it. In Python, abstraction is achieved using abstract classes (via the abc module) and interfaces.
Encapsulation refers to binding data and methods together within a class and restricting direct access to some of the object's components. It focuses on data hiding and protection, typically using private (__) or protected (_) attributes. Encapsulation is about how data is stored and protected, while Abstraction is about how the system is presented to the user.


Q5. What are dunder methods in Python?

Answer:
Dunder methods (also called magic methods or special methods) are methods in Python that have double underscores before and after their names (e.g., __init__, __str__, __repr__, __len__, __add__). The term "dunder" comes from "double underscore."
These methods allow Python objects to interact with built-in Python operations and functions. For example, __init__ is automatically called when an object is created, __str__ defines what is printed when print() is called on an object, and __add__ defines the behavior of the + operator for custom objects. Dunder methods make user-defined classes behave like built-in types.


Q6. Explain the concept of inheritance in OOP.

Answer:
Inheritance is the OOP mechanism by which one class (called a child or derived class) acquires the properties and behaviors (attributes and methods) of another class (called a parent or base class). It promotes code reusability and establishes an "IS-A" relationship between classes.
Python supports multiple types of inheritance: Single Inheritance (one parent), Multiple Inheritance (two or more parents), Multilevel Inheritance (chain of classes), and Hierarchical Inheritance (one parent, multiple children). Using inheritance, the child class can use the parent's methods as-is, override them, or extend them. The super() function is used to call the parent class methods from the child class.


Q7. What is polymorphism in OOP?

Answer:
Polymorphism means "many forms." In OOP, it refers to the ability of different classes to provide a different implementation of the same method name. It allows one interface to be used for a general class of actions, with the specific action determined by the type of the object.
In Python, polymorphism is demonstrated through method overriding (child class redefines parent's method), duck typing (Python checks behavior rather than type), and operator overloading (same operator works differently for different types). For example, the + operator adds two integers but concatenates two strings — this is polymorphism at the operator level.


Q8. How is encapsulation achieved in Python?

Answer:
In Python, encapsulation is achieved by controlling access to class attributes using naming conventions and property decorators:
1. Public attributes (name) – Accessible from anywhere. 2. Protected attributes (_name) – By convention, should not be accessed outside the class or subclasses (single underscore). 3. Private attributes (__name) – Name-mangled by Python to _ClassName__name, making them harder to access from outside (double underscore).
Getter and setter methods or Python's @property decorator are used to safely read/modify private attributes. This ensures data integrity and hides the internal implementation details.


Q9. What is a constructor in Python?

Answer:
A constructor is a special method that is automatically called when an object of a class is created. In Python, the constructor is defined using the __init__() method. It is used to initialize the object's attributes at the time of creation.
The first parameter of __init__() is always self, which refers to the current instance being created. Constructors can accept additional parameters to set initial values. Python also supports a __new__() method that is called before __init__() and is responsible for creating the instance, but __init__() is used in practice for initialization.


Q10. What are class and static methods in Python?

Answer:
Instance methods are the most common type and operate on the instance of the class (using self).
Class methods are methods that receive the class itself as the first argument (cls) rather than an instance. They are defined using the @classmethod decorator. They can access class-level attributes and are commonly used as factory methods.
Static methods do not receive any implicit first argument (no self or cls). They are defined using the @staticmethod decorator. They behave like regular functions but belong to the class's namespace. They are used for utility functions that logically belong to the class but do not need access to instance or class data.


Q11. What is method overloading in Python?

Answer:
Method overloading refers to defining multiple methods with the same name but different parameter signatures. Unlike Java or C++, Python does NOT natively support method overloading in the traditional sense — defining two methods with the same name simply replaces the previous one.
However, Python achieves a similar effect using default arguments, *args, or **kwargs. For example, a method add(a, b=0) can be called as add(5) or add(5, 3), effectively simulating overloading. Python's flexibility with arguments makes traditional method overloading unnecessary, and the same method can handle different input types gracefully.


Q12. What is method overriding in OOP?

Answer:
Method overriding occurs when a child class defines a method with the same name as a method in its parent class. The child's method replaces (overrides) the parent's method when called on a child object. This is a key aspect of runtime polymorphism.
Overriding allows child classes to customize or extend the behavior defined in the parent class. The parent's original method can still be called using super().method_name() if needed. Overriding must have the same method name but can have a different implementation, allowing flexible and extensible code design.


Q13. What is a property decorator in Python?

Answer:
The @property decorator in Python allows a method to be accessed like an attribute, without calling it explicitly as a function. It enables controlled access to private attributes by creating getter, setter, and deleter methods.
When you define a method with @property, it acts as a getter. You can define a setter using @attribute_name.setter and a deleter using @attribute_name.deleter. This approach supports encapsulation while providing a clean interface — users access obj.name instead of obj.get_name(), making the code more Pythonic and intuitive.


Q14. Why is polymorphism important in OOP?

Answer:
Polymorphism is a cornerstone of OOP because it enables writing flexible, reusable, and extensible code. It allows a single interface or method name to work with different types of data or objects, reducing code duplication.
Key advantages include: (1) Code extensibility – new classes can be added with minimal changes to existing code; (2) Maintainability – the same method name can be used across different classes, making code intuitive; (3) Interoperability – objects of different classes can be used interchangeably if they share a common interface; (4) Clean design – it supports the open/closed principle (open for extension, closed for modification).


Q15. What is an abstract class in Python?

Answer:
An abstract class is a class that cannot be instantiated directly and is meant to serve as a blueprint for other classes. It may contain one or more abstract methods — methods that are declared but have no implementation. Subclasses are required to provide concrete implementations for all abstract methods.
In Python, abstract classes are created using the ABC class and @abstractmethod decorator from the abc module. Attempting to instantiate an abstract class or a subclass that hasn't implemented all abstract methods raises a TypeError. Abstract classes are used to enforce a consistent interface across a family of related classes.


Q16. What are the advantages of OOP?

Answer:
Object-Oriented Programming offers numerous advantages:
1. Modularity: Code is organized into classes, making it easier to manage and debug. 2. Reusability: Through inheritance, existing code can be reused without rewriting. 3. Scalability: New features can be added by creating new classes or extending existing ones. 4. Data Security: Encapsulation hides data from direct external access, improving security. 5. Maintainability: Changes to one class rarely affect others, making updates easier. 6. Abstraction: Complexity is hidden, and only relevant details are exposed. 7. Real-world modeling: OOP naturally maps to real-world entities and relationships.


Q17. What is the difference between a class variable and an instance variable?

Answer:
Class Variables are shared among all instances of the class. They are defined directly inside the class body (outside any method) and accessed using the class name or self. Changes to a class variable affect all objects unless overridden at the instance level.
Instance Variables are unique to each object. They are defined inside methods (typically __init__) using self.variable_name. Each object maintains its own copy of instance variables, independent of other objects. If you change an instance variable on one object, it does not affect other objects of the same class.


Q18. What is multiple inheritance in Python?

Answer:
Multiple inheritance is when a class inherits from more than one parent class. Python supports this feature, which allows a child class to combine functionality from multiple base classes.
Python uses the Method Resolution Order (MRO) – specifically the C3 Linearization algorithm – to determine the order in which parent classes are searched when calling a method. You can view the MRO using ClassName.__mro__ or ClassName.mro(). While powerful, multiple inheritance can lead to the "diamond problem," which Python resolves through MRO. It is used carefully to combine unrelated class functionalities.


Q19. Explain the purpose of __str__ and __repr__ methods in Python.

Answer:
Both __str__ and __repr__ are dunder methods that define string representations of an object, but they serve different purposes:
__str__ is meant to provide a human-readable, informal string representation of the object. It is called when you use print(obj) or str(obj). It should be easy to understand for end-users.
__repr__ is meant to provide an official, unambiguous string representation that ideally can be used to recreate the object. It is called when you type the object in the interactive shell or use repr(obj). If __str__ is not defined, Python falls back to __repr__. As a rule of thumb: __str__ is for users, __repr__ is for developers.


Q20. What is the significance of the super() function in Python?

Answer:
The super() function returns a proxy object that delegates method calls to a parent or sibling class. It is most commonly used in the child class's __init__() to call the parent class's constructor, ensuring that the parent's attributes are properly initialized before the child's.
Without super(), you would have to explicitly call the parent class by name (e.g., Animal.__init__(self)), which breaks if the class hierarchy changes. super() makes code more maintainable, particularly in multiple inheritance scenarios where it ensures all parent constructors are called in the correct MRO order.


Q21. What is the significance of the __del__ method in Python?

Answer:
The __del__ method, also called the destructor, is automatically called when an object is about to be destroyed (i.e., when it goes out of scope or is explicitly deleted using del keyword). It is used for cleanup operations, such as closing file handles, releasing database connections, or freeing external resources.
However, the exact timing of when __del__ is called is not guaranteed in Python due to the garbage collector. Python uses reference counting for memory management, and __del__ is invoked when the reference count of an object drops to zero. It should be used carefully and sparingly, as overuse can complicate garbage collection.


Q22. What is the difference between @staticmethod and @classmethod in Python?

Answer:
Both @staticmethod and @classmethod are decorators that modify how methods behave, but they differ in what they receive as their first argument:
@staticmethod: Does not receive any implicit first argument. It behaves like a regular function inside the class namespace. It cannot access or modify class state or instance state. Used for utility functions that logically belong to the class.
@classmethod: Receives the class itself (cls) as the first argument. It can access and modify class-level attributes. It is often used to create alternative constructors (factory methods). Example: Date.from_string("2024-01-15") as a classmethod that creates a Date object from a string.


Q23. How does polymorphism work in Python with inheritance?

Answer:
Polymorphism with inheritance in Python works through method overriding. When a parent class defines a method, child classes can override it with their own implementation. When you call the method on an object, Python dynamically determines which version to run based on the actual type of the object at runtime — this is called runtime polymorphism or dynamic dispatch.
For example, if Animal has a speak() method, and Dog and Cat both override it, calling animal_object.speak() will invoke the correct version depending on whether the object is a Dog or Cat. This allows writing generic code (e.g., looping over a list of Animal objects) while getting type-specific behavior automatically.


Q24. What is method chaining in Python OOP?

Answer:
Method chaining is a programming technique where multiple methods are called on the same object in a single line, with each method call returning the object itself (self). This allows for a fluent interface and makes the code more concise and readable.
To enable method chaining, each method must return self at the end. For example: obj.set_name("Alice").set_age(25).display() — here each method returns the object, so the next method can be called immediately. Method chaining is widely used in builder patterns, query builders (like Django ORM), and data manipulation libraries like pandas.


Q25. What is the purpose of the __call__ method in Python?

Answer:
The __call__ method allows an instance of a class to be called like a regular function. When you define __call__ in a class, you can use the object as if it were a callable (function). This is triggered when the object is followed by parentheses: obj().
This is useful for creating function-like objects (functors), decorators, and memoization wrappers. For example, a Multiplier class with __call__ defined can work as: double = Multiplier(2); double(5) → 10. The __call__ method makes objects flexible and allows state to be maintained across multiple function-like invocations.



Part B: Practical Questions – Python OOP Programs
The following section contains 18 complete Python programs with detailed explanations, logic breakdowns, and sample inputs/outputs.

Q1. Create a parent class Animal with a method speak() that prints a generic message. Create a child class Dog that overrides the speak() method to print 'Bark!'.

Complete Python Code:
class Animal:
    def speak(self):
        print('The animal makes a sound.')

class Dog(Animal):
    def speak(self):
        print('Bark!')

# Creating objects
generic_animal = Animal()
my_dog = Dog()

generic_animal.speak()   # Calls Animal's speak()
my_dog.speak()           # Calls Dog's overridden speak()

Code Explanation:
The Animal class defines a generic speak() method. The Dog class inherits from Animal and overrides speak() with a specific implementation. When speak() is called on a Dog object, Python uses the child class's version due to method overriding. This demonstrates single inheritance and runtime polymorphism.

Functions & Methods Used:
speak() method in Animal (base), speak() in Dog (overridden), Object instantiation with ()

Sample Output:
The animal makes a sound.
Bark!

Advantage / Application:
Enables different animal types to have unique sounds while sharing a common interface.


Q2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.

Complete Python Code:
from abc import ABC, abstractmethod
import math

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return round(math.pi * self.radius ** 2, 2)

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width

c = Circle(7)
r = Rectangle(4, 5)
print(f'Circle Area: {c.area()}')
print(f'Rectangle Area: {r.area()}')

Code Explanation:
Shape is an abstract class that cannot be instantiated. It forces any derived class to implement the area() method using @abstractmethod. Circle calculates area using πr², while Rectangle uses length × width. This demonstrates abstraction — the user knows what to call (area()), not how it's calculated internally.

Functions & Methods Used:
ABC from abc module, @abstractmethod decorator, math.pi for π constant, area() implemented differently in each subclass

Sample Output:
Circle Area: 153.94
Rectangle Area: 20

Advantage / Application:
Enforces a consistent interface across all shape subclasses. Any new shape must implement area().


Q3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.

Complete Python Code:
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type
    def display(self):
        print(f'Vehicle Type: {self.vehicle_type}')

class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)
        self.brand = brand
    def display(self):
        super().display()
        print(f'Brand: {self.brand}')

class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery_capacity):
        super().__init__(vehicle_type, brand)
        self.battery_capacity = battery_capacity
    def display(self):
        super().display()
        print(f'Battery: {self.battery_capacity} kWh')

ec = ElectricCar('4-Wheeler', 'Tesla', 100)
ec.display()

Code Explanation:
This demonstrates multi-level inheritance: ElectricCar → Car → Vehicle. Each child class calls super().__init__() to initialize parent attributes. super().display() chains the display calls up the hierarchy, so ElectricCar's display shows all three levels of information. The inheritance chain creates an IS-A relationship: ElectricCar IS-A Car, Car IS-A Vehicle.

Functions & Methods Used:
super().__init__() to call parent constructors, super().display() to chain display methods, multi-level __init__ chaining

Sample Output:
Vehicle Type: 4-Wheeler
Brand: Tesla
Battery: 100 kWh

Advantage / Application:
Avoids code duplication across the hierarchy. Each level adds only its own attribute.


Q4. Demonstrate polymorphism by creating a base class Bird with a method fly(). Create two derived classes Sparrow and Penguin that override the fly() method.

Complete Python Code:
class Bird:
    def fly(self):
        print('Some birds can fly.')

class Sparrow(Bird):
    def fly(self):
        print('Sparrow flies swiftly through the sky!')

class Penguin(Bird):
    def fly(self):
        print('Penguins cannot fly — they swim instead!')

# Polymorphic behavior
birds = [Sparrow(), Penguin(), Bird()]
for bird in birds:
    bird.fly()

Code Explanation:
All three classes define a fly() method. When iterating over a list of Bird objects, Python dynamically calls the correct fly() based on each object's actual type. This is runtime polymorphism. The same method name produces different behaviors — Sparrow flies, Penguin swims. This demonstrates duck typing and method overriding working together.

Functions & Methods Used:
fly() overridden in each subclass, iteration with for loop, list of heterogeneous Bird objects

Sample Output:
Sparrow flies swiftly through the sky!
Penguins cannot fly — they swim instead!
Some birds can fly.

Advantage / Application:
Allows processing diverse object types through a uniform interface, enabling scalable code.


Q5. Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.

Complete Python Code:
class BankAccount:
    def __init__(self, owner, initial_balance=0):
        self.owner = owner
        self.__balance = initial_balance   # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f'Deposited ₹{amount}. New Balance: ₹{self.__balance}')
        else:
            print('Invalid deposit amount.')

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f'Withdrawn ₹{amount}. Remaining Balance: ₹{self.__balance}')
        else:
            print('Insufficient funds or invalid amount.')

    def check_balance(self):
        print(f'{self.owner}\'s Balance: ₹{self.__balance}')

acc = BankAccount('Rahul', 1000)
acc.deposit(500)
acc.withdraw(200)
acc.check_balance()

Code Explanation:
__balance is a private attribute (double underscore), preventing direct external access. All modifications go through deposit() and withdraw() which validate the input before updating the balance. This ensures data integrity — no negative deposits or overdrafts. Encapsulation protects the account data and provides a safe public interface.

Functions & Methods Used:
__balance private attribute, deposit() with validation, withdraw() with balance check, check_balance() as public getter

Sample Output:
Deposited ₹500. New Balance: ₹1500
Withdrawn ₹200. Remaining Balance: ₹1300
Rahul's Balance: ₹1300

Advantage / Application:
Prevents accidental or malicious tampering with financial data. Models real-world banking rules.


Q6. Demonstrate runtime polymorphism using a method play() in a base class Instrument. Derive classes Guitar and Piano that implement their own version of play().

Complete Python Code:
class Instrument:
    def play(self):
        print('Playing a musical instrument...')

class Guitar(Instrument):
    def play(self):
        print('Strumming the guitar strings — ♪ Twang!')

class Piano(Instrument):
    def play(self):
        print('Pressing piano keys — ♪ Melody flows!')

# Runtime polymorphism demonstration
instruments = [Guitar(), Piano(), Instrument()]
print('--- Concert Time ---')
for instrument in instruments:
    instrument.play()

Code Explanation:
The Instrument class provides a generic play() method. Guitar and Piano each override it with specific implementations. When play() is called on each item in the list, Python resolves which version to execute at runtime based on the actual object type. This is runtime polymorphism (dynamic dispatch), meaning the decision is made during program execution, not at compile time.

Functions & Methods Used:
play() overridden in subclasses, polymorphic loop, runtime type resolution

Sample Output:
--- Concert Time ---
Strumming the guitar strings — ♪ Twang!
Pressing piano keys — ♪ Melody flows!
Playing a musical instrument...

Advantage / Application:
Easily extensible: add a new instrument class without changing the loop logic.


Q7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.

Complete Python Code:
class MathOperations:
    total_operations = 0   # Class variable

    @classmethod
    def add_numbers(cls, a, b):
        cls.total_operations += 1
        result = a + b
        print(f'{a} + {b} = {result}')
        return result

    @staticmethod
    def subtract_numbers(a, b):
        result = a - b
        print(f'{a} - {b} = {result}')
        return result

MathOperations.add_numbers(10, 5)
MathOperations.subtract_numbers(10, 5)
print(f'Total add operations: {MathOperations.total_operations}')

Code Explanation:
add_numbers is a @classmethod — it receives cls (the class itself) and can access/modify class variables like total_operations. subtract_numbers is a @staticmethod — it receives no implicit argument and simply performs a utility calculation. Both can be called on the class directly without creating an instance. The class variable tracks total additions across all calls.

Functions & Methods Used:
@classmethod with cls parameter, @staticmethod with no special parameter, class variable total_operations

Sample Output:
10 + 5 = 15
10 - 5 = 5
Total add operations: 1

Advantage / Application:
Class and static methods provide namespace organization without requiring object instantiation.


Q8. Implement a class Person with a class method to count the total number of persons created.

Complete Python Code:
class Person:
    person_count = 0   # Class variable shared by all instances

    def __init__(self, name, age):
        self.name = name
        self.age = age
        Person.person_count += 1   # Increment on each object creation

    @classmethod
    def get_count(cls):
        return cls.person_count

    def display(self):
        print(f'Name: {self.name}, Age: {self.age}')

p1 = Person('Alice', 25)
p2 = Person('Bob', 30)
p3 = Person('Charlie', 22)

p1.display()
p2.display()
p3.display()
print(f'Total persons created: {Person.get_count()}')

Code Explanation:
person_count is a class variable initialized to 0. Each time __init__ runs (i.e., a new Person is created), person_count is incremented. The @classmethod get_count() provides a clean interface to retrieve this count. This pattern is commonly used for singleton patterns, instance tracking, and resource management.

Functions & Methods Used:
__init__ increments class counter, @classmethod get_count() reads class state, class variable person_count

Sample Output:
Name: Alice, Age: 25
Name: Bob, Age: 30
Name: Charlie, Age: 22
Total persons created: 3

Advantage / Application:
Efficiently tracks instance counts without needing global variables or external counters.


Q9. Write a class Fraction with attributes numerator and denominator. Override the __str__ method to display the fraction as 'numerator/denominator'.

Complete Python Code:
import math

class Fraction:
    def __init__(self, numerator, denominator):
        if denominator == 0:
            raise ValueError('Denominator cannot be zero!')
        common = math.gcd(abs(numerator), abs(denominator))
        self.numerator = numerator // common
        self.denominator = denominator // common

    def __str__(self):
        return f'{self.numerator}/{self.denominator}'

    def __repr__(self):
        return f'Fraction({self.numerator}, {self.denominator})'

    def __add__(self, other):
        new_num = self.numerator * other.denominator + other.numerator * self.denominator
        new_den = self.denominator * other.denominator
        return Fraction(new_num, new_den)

f1 = Fraction(1, 2)
f2 = Fraction(2, 4)
f3 = f1 + f2
print(f'f1 = {f1}')
print(f'f2 = {f2}')
print(f'f1 + f2 = {f3}')

Code Explanation:
The Fraction class simplifies fractions using GCD in __init__. __str__ returns a human-readable 'numerator/denominator' string used by print(). __repr__ returns a developer-friendly representation. __add__ enables the + operator between fractions. math.gcd ensures fractions are always stored in simplest form.

Functions & Methods Used:
math.gcd() for simplification, __str__ for display, __repr__ for repr(), __add__ for + operator overloading

Sample Output:
f1 = 1/2
f2 = 1/2
f1 + f2 = 1/1

Advantage / Application:
Demonstrates operator overloading and dunder methods for creating Pythonic custom types.


Q10. Demonstrate operator overloading by creating a class Vector and overriding the __add__ method to add two vectors.

Complete Python Code:
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y)

    def __str__(self):
        return f'Vector({self.x}, {self.y})'

    def magnitude(self):
        return round((self.x**2 + self.y**2) ** 0.5, 2)

v1 = Vector(3, 4)
v2 = Vector(1, 2)
v3 = v1 + v2
v4 = v1 - v2
print(f'v1 = {v1}, Magnitude = {v1.magnitude()}')
print(f'v2 = {v2}')
print(f'v1 + v2 = {v3}')
print(f'v1 - v2 = {v4}')

Code Explanation:
__add__ is called when the + operator is used between two Vector objects. It creates and returns a new Vector whose x and y are the sums of the original vectors' components. __sub__ similarly handles subtraction. __str__ provides readable output. magnitude() computes the Euclidean length using the Pythagorean theorem.

Functions & Methods Used:
__add__ overloads +, __sub__ overloads -, __str__ for print output, magnitude() for Euclidean distance

Sample Output:
v1 = Vector(3, 4), Magnitude = 5.0
v2 = Vector(1, 2)
v1 + v2 = Vector(4, 6)
v1 - v2 = Vector(2, 2)

Advantage / Application:
Allows mathematical types to use natural Python operators, making code intuitive and readable.


Q11. Create a class Person with attributes name and age. Add a method greet() that prints 'Hello, my name is {name} and I am {age} years old.'

Complete Python Code:
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f'Hello, my name is {self.name} and I am {self.age} years old.')

    def birthday(self):
        self.age += 1
        print(f'Happy Birthday {self.name}! You are now {self.age} years old.')

    def __str__(self):
        return f'Person(name={self.name}, age={self.age})'

p1 = Person('Rahul', 21)
p2 = Person('Priya', 19)
p1.greet()
p2.greet()
p1.birthday()
print(p1)

Code Explanation:
The Person class stores name and age as instance attributes set via __init__. The greet() method uses an f-string to produce a personalized message. birthday() increments age, showing how methods can modify instance state. __str__ provides a clean string representation for print(). This is a foundational OOP class design example.

Functions & Methods Used:
greet() with f-string, birthday() modifies self.age, __str__ for display, __init__ for initialization

Sample Output:
Hello, my name is Rahul and I am 21 years old.
Hello, my name is Priya and I am 19 years old.
Happy Birthday Rahul! You are now 22 years old.
Person(name=Rahul, age=22)

Advantage / Application:
Demonstrates basic class design, instance attributes, and method interactions on shared state.


Q12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades.

Complete Python Code:
class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades   # List of grades

    def average_grade(self):
        if not self.grades:
            return 0
        return round(sum(self.grades) / len(self.grades), 2)

    def highest_grade(self):
        return max(self.grades)

    def lowest_grade(self):
        return min(self.grades)

    def display(self):
        print(f'Student: {self.name}')
        print(f'Grades: {self.grades}')
        print(f'Average Grade: {self.average_grade()}')
        print(f'Highest: {self.highest_grade()} | Lowest: {self.lowest_grade()}')

s = Student('Anjali', [85, 92, 78, 95, 88])
s.display()

Code Explanation:
The Student class accepts grades as a list. average_grade() uses sum() and len() to compute the average and round() to limit decimal places. highest_grade() and lowest_grade() use Python's built-in max() and min() functions. The guard clause (if not self.grades) prevents ZeroDivisionError for empty grade lists.

Functions & Methods Used:
average_grade() with sum()/len(), max() and min() for highest/lowest, round() for precision, empty list guard

Sample Output:
Student: Anjali
Grades: [85, 92, 78, 95, 88]
Average Grade: 87.6
Highest: 95 | Lowest: 78

Advantage / Application:
Models real academic records; easily extensible with letter grade or pass/fail methods.


Q13. Create a class Rectangle with methods set_dimensions() to set the dimensions and area() to calculate the area.

Complete Python Code:
class Rectangle:
    def __init__(self):
        self.length = 0
        self.width = 0

    def set_dimensions(self, length, width):
        if length <= 0 or width <= 0:
            print('Dimensions must be positive.')
            return
        self.length = length
        self.width = width
        print(f'Dimensions set: {length} x {width}')

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

    def is_square(self):
        return self.length == self.width

rect = Rectangle()
rect.set_dimensions(6, 4)
print(f'Area: {rect.area()}')
print(f'Perimeter: {rect.perimeter()}')
print(f'Is Square: {rect.is_square()}')

Code Explanation:
The Rectangle is initialized with zero dimensions. set_dimensions() validates inputs to ensure positive values before updating attributes. area() returns length × width. Additional utility methods perimeter() and is_square() enhance the class. Input validation in set_dimensions() prevents logically incorrect rectangles.

Functions & Methods Used:
set_dimensions() with validation, area() calculation, perimeter() utility, is_square() boolean check

Sample Output:
Dimensions set: 6 x 4
Area: 24
Perimeter: 20
Is Square: False

Advantage / Application:
Separation of initialization and dimension-setting allows flexible object creation and reuse.


Q14. Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.

Complete Python Code:
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

    def display(self):
        print(f'Employee: {self.name}')
        print(f'Salary: ₹{self.calculate_salary()}')

class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)
        self.bonus = bonus

    def calculate_salary(self):
        base = super().calculate_salary()
        return base + self.bonus

    def display(self):
        print(f'Manager: {self.name}')
        print(f'Base Salary: ₹{super().calculate_salary()}')
        print(f'Bonus: ₹{self.bonus}')
        print(f'Total Salary: ₹{self.calculate_salary()}')

emp = Employee('Ramesh', 160, 250)
mgr = Manager('Suresh', 160, 400, 5000)
emp.display()
print('---')
mgr.display()

Code Explanation:
Employee calculates base salary as hours × rate. Manager inherits from Employee and overrides calculate_salary() to add a bonus. super().calculate_salary() reuses the parent's logic without duplication. Manager.display() shows a breakdown of base salary and bonus. This models real-world employee payroll systems using inheritance.

Functions & Methods Used:
calculate_salary() in Employee (base), overridden in Manager, super() to reuse parent method, display() at both levels

Sample Output:
Employee: Ramesh
Salary: ₹40000
---
Manager: Suresh
Base Salary: ₹64000
Bonus: ₹5000
Total Salary: ₹69000

Advantage / Application:
Clean salary hierarchy; adding new employee types (Intern, Director) requires minimal changes.


Q15. Create a class Product with attributes name, price, and quantity. Implement a method total_price() that calculates the total price of the product.

Complete Python Code:
class Product:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        return self.price * self.quantity

    def apply_discount(self, discount_percent):
        discounted = self.price * (1 - discount_percent / 100)
        return round(discounted * self.quantity, 2)

    def __str__(self):
        return (f'Product: {self.name} | Price: ₹{self.price} | '
                f'Qty: {self.quantity} | Total: ₹{self.total_price()}')

p1 = Product('Laptop', 55000, 3)
p2 = Product('Mouse', 800, 10)
print(p1)
print(p2)
print(f'Laptop with 10% discount total: ₹{p1.apply_discount(10)}')

Code Explanation:
Product stores name, price, and quantity. total_price() multiplies price by quantity to get the total cost. apply_discount() calculates the final total after a percentage discount, useful in shopping cart scenarios. __str__ formats the product details cleanly for display. This models a simple e-commerce product entity.

Functions & Methods Used:
total_price() = price × quantity, apply_discount() with percentage calculation, __str__ for display

Sample Output:
Product: Laptop | Price: ₹55000 | Qty: 3 | Total: ₹165000
Product: Mouse | Price: ₹800 | Qty: 10 | Total: ₹8000
Laptop with 10% discount total: ₹148500.0

Advantage / Application:
Reusable product model for inventory, e-commerce, or billing systems.


Q16. Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method.

Complete Python Code:
from abc import ABC, abstractmethod

class Animal(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def sound(self):
        pass

    def describe(self):
        print(f'I am {self.name} and I say: ', end='')
        self.sound()

class Cow(Animal):
    def sound(self):
        print('Moo!')

class Sheep(Animal):
    def sound(self):
        print('Baa!')

animals = [Cow('Bessie'), Sheep('Dolly')]
for animal in animals:
    animal.describe()

Code Explanation:
Animal is an abstract base class with an abstract sound() method that forces subclasses to implement it. Cow and Sheep each provide their specific sound. The describe() method in the base class calls self.sound(), which uses polymorphism to call the right subclass method. Note: Animal() cannot be instantiated directly.

Functions & Methods Used:
@abstractmethod forces implementation in subclasses, describe() uses polymorphic self.sound(), ABC prevents direct instantiation

Sample Output:
I am Bessie and I say: Moo!
I am Dolly and I say: Baa!

Advantage / Application:
Guarantees all Animal subclasses have a sound() method, enforcing a contract across the hierarchy.


Q17. Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that returns a formatted string with the book's details.

Complete Python Code:
class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        return (f'Title    : {self.title}\n'
                f'Author   : {self.author}\n'
                f'Published: {self.year_published}')

    def age(self):
        from datetime import date
        return date.today().year - self.year_published

    def __str__(self):
        return self.get_book_info()

    def __repr__(self):
        return f'Book({self.title!r}, {self.author!r}, {self.year_published})'

b1 = Book('The Alchemist', 'Paulo Coelho', 1988)
b2 = Book('Python Crash Course', 'Eric Matthes', 2015)
print(b1.get_book_info())
print('---')
print(b2)
print(f'Age of book: {b2.age()} years')

Code Explanation:
Book stores three attributes set in __init__. get_book_info() formats them into a multi-line string using an f-string with escape sequences for alignment. age() calculates how old the book is using datetime. __str__ delegates to get_book_info(), so print(book) shows formatted details. __repr__ provides a developer-friendly representation.

Functions & Methods Used:
get_book_info() for formatted output, age() with datetime, __str__ delegates to get_book_info(), __repr__ for debugging

Sample Output:
Title    : The Alchemist
Author   : Paulo Coelho
Published: 1988
---
Title    : Python Crash Course
Author   : Eric Matthes
Published: 2015
Age of book: 11 years

Advantage / Application:
Models a library system record; easily extendable with ISBN, genre, and availability attributes.


Q18. Create a class House with attributes address and price. Create a derived class Mansion that adds an attribute number_of_rooms.

Complete Python Code:
class House:
    def __init__(self, address, price):
        self.address = address
        self.price = price

    def display(self):
        print(f'Address : {self.address}')
        print(f'Price   : ₹{self.price:,}')

    def price_per_sqft(self, area_sqft):
        return round(self.price / area_sqft, 2)

class Mansion(House):
    def __init__(self, address, price, number_of_rooms):
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms

    def display(self):
        super().display()
        print(f'Rooms   : {self.number_of_rooms}')
        print(f'Type    : Mansion')

h = House('12 Baker Street, Delhi', 5000000)
m = Mansion('1 Royal Avenue, Mumbai', 50000000, 15)
print('--- House ---')
h.display()
print('\n--- Mansion ---')
m.display()
print(f'Price per sqft (5000 sqft): ₹{m.price_per_sqft(5000)}')

Code Explanation:
House is the base class with address and price. Mansion inherits from House and adds number_of_rooms. super().__init__() initializes the parent's attributes. Mansion.display() calls super().display() first, then adds room information. The price_per_sqft() utility method is inherited by Mansion without any duplication.

Functions & Methods Used:
super().__init__() for parent initialization, super().display() for chained display, price_per_sqft() inherited utility, f-string with :, formatting

Sample Output:
--- House ---
Address : 12 Baker Street, Delhi
Price   : ₹5,000,000

--- Mansion ---
Address : 1 Royal Avenue, Mumbai
Price   : ₹50,000,000
Rooms   : 15
Type    : Mansion
Price per sqft (5000 sqft): ₹10000.0

Advantage / Application:
Models a real estate hierarchy; easily extensible with Apartment, Villa, or Penthouse subclasses.
```
